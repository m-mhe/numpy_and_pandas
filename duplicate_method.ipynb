# Using duplicated() method to find out column values is unique or not

```python
import pandas as pd
```

```python
dfCmt = pd.read_csv("laptop_price.csv", encoding = "unicode_escape")
dfCmt
```

```python
dfCmt[~dfCmt.duplicated("laptop_ID")].value_counts("laptop_ID") # '~' means not in pandas
```

```python
dfCmt.set_index("laptop_ID", inplace = True) # So a non duplicate field can be index.
dfCmt
```

# Finding duplicates from multiple columns

```python
dfCmt[dfCmt.duplicated(["Memory", "Cpu"])].sort_values(["Cpu","Memory"])
```

```python
dfCmt[~dfCmt.duplicated(["Memory", "Cpu"])].sort_values(["Cpu","Memory"]) # '~' means not in pandas
```

# Most & Least expansive from each company

```python
dfCmt.sort_values("Price_euros", inplace = True)
dfCmt
```

```python
dfLep = dfCmt.duplicated("Company", keep = "first") # If we set keep "first"
dfLep = dfCmt[~dfLep]
dfLep # Least Expensive Product (LEP) from each company.
```

```python
dfMep = dfCmt[~dfCmt.duplicated("Company", keep = "last")] # If we set keep "last"
dfMep # Most Expensive Product (MEP) from each company.
```

```python
dfCmt[~dfCmt.duplicated("Company", keep = False)] # If we set keep False
```
