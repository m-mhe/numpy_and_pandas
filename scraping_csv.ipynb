# Initializing

```python
import pandas as pd
import csv

baseURL = "https://www.football-data.co.uk/mmz4281/"
spanClubFootballDataFrame = []
```

# Formatting URLs

```python
# print(str(1).zfill(2)) for getting leading 0.

for year in range(0,25):
    season = str(year).zfill(2)+str(year+1).zfill(2)
    for league in range(1,3):
        print(f"{baseURL}{season}/SP{league}.csv")
        #tDf = pd.read_csv(f"{baseURL}{season}/SP{league}.csv")
        
```

```python
for year in range(0,25):
    season = str(year).zfill(2)+str(year+1).zfill(2)
    for league in range(1,3):
        tDf = pd.read_csv(f"{baseURL}{season}/SP{league}.csv",
                         quoting=csv.QUOTE_MINIMAL,  # or QUOTE_ALL, depending on your file
                         engine="python",on_bad_lines="warn",    # or "skip", drops any row that doesnt match the column count
                         encoding="utf-8"
                         ) # encoding = "unicode_escape"
        tDf.rename(columns = {"FTHG":"HomeTeamGoal","FTAG":"AwayTeamGoal"}, inplace = True)
        spanClubFootballDataFrame.append(tDf)
```
