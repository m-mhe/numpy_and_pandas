# Selecting one or multiple columns

```python
import pandas as pd
import numpy as np

nameList = ["Emon", "Upal", "Habib", "Sufia", "Supty", "Faria", "Afrin","Tanvir", "Jasim"]
rollList = np.random.randint(242001300, 242001323, size = len(nameList)) # It genarates a list of random integer.
dsOne = {"Name":nameList, "Roll":list(rollList)} # or we can write like this: "dsOne = {"Name":nameList, "Roll":rollList}".
dfOne = pd.DataFrame(dsOne)
print(dfOne)

print(dfOne["Name"]) # Selecting one column, since it's one column the return type will be a series.

totalScoreList = np.random.uniform(35,101, size = len(nameList))
dfOne["Score"] = totalScoreList # Adding new column name sore.
print(round(dfOne,2)) # It rounds float values in 2 dp.

round(dfOne[["Name", "Score"]],2) # Selecting multiple columns, since it's multiple columns the return type will be a dataframe.
```

# Adding a new column

```python
arr = np.arange(1,len(nameList)+1)
dfOne["Age"] = arr
dfOne
```

# Editing a column

```python
dfOne["Age"] = 22 # all values of column Age will set as 22 (scalar value)
dfOne
```

```python
banglaScore = np.random.randint(60,101, len(nameList)) # Generates 9 random numbers from 22 to 25.
dfOne["Bangla"] = banglaScore
round(dfOne, 2)
```

# Deleting multiple columns

```python
dfOne.drop(columns=["Age", "Score"], inplace = True) # inplace true means the changes will get apply on original dataframe.
dfOne
```

# Adding using assign() method

```python
eS = list(np.random.randint(45,101, size = len(nameList)))
mS = np.random.randint(10,101, len(nameList))
dsS = np.random.randint(25,101,len(nameList))
cpS = np.random.randint(5,101,len(nameList))

sE = pd.Series(eS, index = np.arange(0,len(nameList)))
sM = pd.Series(mS)
sD = pd.Series(dsS)
sC = pd.Series(cpS)

dfOne = dfOne.assign(Data_Structure = dsS, C_Programming = sC,English = sE, Math = sM)
dfOne
```

# Adding using insert() method

```python
age = np.random.randint(22,25,len(nameList))
dfOne.insert(2, "Age", age) # 2 is index number, "Age" is column name, age is our data
dfOne
```
