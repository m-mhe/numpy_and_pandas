# Create a column based on one condition uasing np.where():

```python
import pandas as pd
import numpy as np
```

```python
dfLaptops = pd.read_csv("laptop_price.csv", encoding = "unicode_escape")
dfLaptops
```

```python
np.where(dfLaptops["Inches"]>15) # Returns index where screen size is bigger than 15 inches.
```

```python
np.where(dfLaptops["Inches"]>=15, "Big Screen", "Small Screen") # Returns an array based on screen size.
```

```python
dfLaptops = dfLaptops.assign(Display = np.where(dfLaptops["Inches"]>=15, "Big Screen", "Small Screen")) # Creating a column based on one condition uasing np.where():
dfLaptops.head(5)
```

```python
dfLaptops["Display"].value_counts()
```
