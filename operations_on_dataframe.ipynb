# Genarel operations on a dataframe

```python
import pandas as pd
import numpy as np
```

```python
# Preparing dataframe.
dfResult = pd.read_csv("result.csv")
gL = ['male','male','male','female','female','female','female','male','female',]
dfResult.drop(columns=["Unnamed: 0"], inplace = True)
dfResult.insert(1,"Gender", gL)
dfResult
```

```python
print(dfResult["Bangla"].sum()) # Getting total value of a column
print(dfResult["Bangla"].count()) # Getting total row of a column
print(dfResult["Bangla"].mean().round(2)) # Getting avarage value of a column
print(dfResult["Bangla"].std()) # Getting standard deviation of a column
print(dfResult["Bangla"].min())
print(dfResult["Bangla"].max())
dfResult.describe()
```

```python
print(dfResult["Bangla"]+dfResult["Data_Structure"]+dfResult["C_Programming"]) # Sum of values from row.
```

```python
dfResult["Avg_Number"] = ((dfResult["Bangla"]+dfResult["Data_Structure"]+dfResult["C_Programming"])/3) # Doing basic math from values of row.
dfResult
```

```python
dfResult["GPA"] = ((dfResult["Bangla"]/23.25+dfResult["Data_Structure"]/23.25+dfResult["C_Programming"]/23.25+dfResult["English"]/23.25+dfResult["Math"]/23.25)/5) # Doing basic math from values of row.
dfResult
```

```python
dfResult["Gender"].value_counts() # Checking howmany male and female are in our dataframe.
```

```python
dfResult["Gender"].value_counts(normalize = True) # Checking howmany male and female are in our dataframe in percentage form.
```
