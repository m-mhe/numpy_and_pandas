# Smallest and bigest screen using drop_duplicates() method

```python
import pandas as pd

dfCmt = pd.read_csv("laptop_price.csv", encoding = "unicode_escape")
dfCmt
```

```python
dfS = dfCmt[["Company","Inches", "Price_euros"]]
dfS
```

```python
dfS.sort_values("Inches", inplace = True)
dfS
```

```python
dfSsS = dfS.drop_duplicates("Company", keep = "first")
dfSsS # Smallest
```

```python
dfBsS = dfS.drop_duplicates("Company", keep = "last")
dfBsS
```
