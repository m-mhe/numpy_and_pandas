# Creating DataFrame

### Using Numpy

```python
import numpy as np
import pandas as pd

# Using Numpy
dataOne = np.array([["Emon",242001322],["Upal",46574654],["Irfan",737654]]) # Creating an 2D array
dfOne = pd.DataFrame(dataOne, columns = ["Name", "Roll"])
print("\n")
print(dfOne)
print("\n")

# Using Dictionary
names = []
roll = []
i = 0
while i<=1:
    inp = input(f"Enter {i} name:")
    names.append(inp)
    inp = input(f"Enter {i} roll:")
    roll.append(inp)
    i += 1
dataTwo = {"Name":names,"Roll":roll}
dfTwo = pd.DataFrame(dataTwo)
print("\n")
print(dfTwo)
print("\n")

# Reading CSV
dfThree = pd.read_csv("dummy.csv")
print(dfThree.head()) # Returns first 5 rows
print("\n")
dfThree
```
