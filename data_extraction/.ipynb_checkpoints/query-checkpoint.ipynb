# query()

```python
import pandas as pd

df = pd.read_csv("players_20.csv")
df.head(3)
```

```python
df["dob"] = pd.to_datetime(df["dob"])
df.head(2)
```

```python
df.query("(dob.dt.year>2000 or age == 32) and (height_cm > 202 or height_cm == 170)") # SO as we see, this is same as boolean slicing.
```
