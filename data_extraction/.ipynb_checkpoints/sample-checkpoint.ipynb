# Sample

```python
import pandas as pd
import numpy as np
```

```python
dfPlayers = pd.read_csv("players_20.csv")
dfPlayers.head(3)
```

```python
dfPlayers.sample(20) #To get 20 random samples from dataframe
```

```python
dfPlayers.sample(3, random_state = 420) # Now randomization will happen based on this number 420, if we add this number then we will get this fixed random values.
```

```python
dfPlayers.sample(4, random_state = 420) # Those fixed three row + one.
```

```python
dfPlayers.sample(frac = 0.01) # To get 1% of data from dataframe
```

```python
dfPlayers.sample(frac = 9, replace = True) # To get 900% upsample.
```
