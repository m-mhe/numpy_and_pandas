# Settingup data frame

```python
import pandas as pd

dfPlayers = pd.read_csv("players_20.csv")
dfPlayers.head(3)
```

```python
dfPlayers[dfPlayers.duplicated("short_name")]
```

```python
dfPlayers.drop_duplicates("short_name", keep = "first", inplace = True)
dfPlayers[dfPlayers.duplicated("short_name")] # Hence no duplicated values found in the column, we can make it index.
```

```python
dfPlayers.set_index("short_name", inplace = True)
dfPlayers
```

```python
dfPlayers.rename(columns = {"dob":"date_of_birth"}, inplace = True)
dfPlayers.head(3)
```

```python
dfPlayers = dfPlayers[["overall","long_name","date_of_birth","age","height_cm","nationality","club"]]
dfPlayers.head(3)
```

```python
dfPlayers.rename(columns = {"overall":"overall_rating"}, inplace = True)
```

```python
dfPlayers.head(10)
```

# Selecting with .loc() method

```python
dfPlayers.loc["K. De Bruyne", "overall_rating"] # Selecting a perticular data cell
```

```python
dfPlayers.loc[:, "overall_rating"] # To get all rows under a column
```

```python
dfPlayers.loc[:, "L. Messi"] # This method doesn't work on index column name.
```

```python
dfPlayers.loc["L. Messi",:] # To get all columns under a row index.
```

```python
dfPlayers.loc["L. Messi"] # To get all columns under a row index.
```

```python
dfPlayers.loc[["L. Messi", "Neymar Jr","Cristiano Ronaldo"]] # To get all columns under multiple row index.
```

```python
dfPlayers.loc[["overall_rating", "club"]] # This doesn't work!
```

```python
dfPlayers.loc[["L. Messi", "Neymar Jr","Cristiano Ronaldo"], ["overall_rating", "club"]] # Selecting using list of values.
```

```python
dfPlayers.loc[["L. Messi", "Neymar Jr","Cristiano Ronaldo"], "club"]
```

```python
dfPlayers.loc[:,["overall_rating", "club"]]
```

```python
dfPlayers.loc[["L. Messi", "Neymar Jr","Cristiano Ronaldo"],:]
```

# Selecting range of data with slice

```python
dfPlayers.loc[["L. Messi", "Neymar Jr","Cristiano Ronaldo"],"overall_rating":"nationality"] # Slicing columns
```

```python
dfPlayers.loc["L. Messi":"E. Hazard", ["overall_rating","nationality"]] # Slicing rows with row index
```

```python
dfPlayers.iloc[0:10, 0:2] # Slicing botyh using .iloc(), we can do the same with .loc() also.
```
